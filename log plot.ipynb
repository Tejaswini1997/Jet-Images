```python
import numpy as np

def translate_jet_image(jet_image):
    # Get the shape of the image array
    n_eta, n_phi = jet_image.shape

    # Compute the eta and phi values for each pixel in the image
    eta, phi = np.meshgrid(np.arange(n_eta), np.arange(n_phi), indexing='ij')

    # Find the indices of the pixel with the highest intensity
    max_idx = np.unravel_index(np.argmax(jet_image), jet_image.shape)

    # Compute the translation values to move the highest intensity pixel to the center
    d_eta = max_idx[0] - n_eta//2
    d_phi = max_idx[1] - n_phi//2

    # Compute the Lorentz factor for the boost
    gamma = np.cosh(d_eta)

    # Compute the transverse energy for each pixel
    pt_image = jet_image / gamma / np.cosh(eta - d_eta)

    # Roll the image array by the translation values
    translated_image = np.roll(pt_image, (-d_eta, -d_phi), axis=(0,1))

    return translated_image
```

```python
def make_bins(df, x_bin, y_bin):

    X, Y = np.arange(df['del_eta'].min(),df['del_eta'].max()+x_bin, x_bin), np.arange(df['del_phi'].min(),df['del_phi'].max()+y_bin, y_bin)
    df['xbin'] = pd.cut(df['del_eta'], X)
    df['ybin'] = pd.cut(df['del_phi'], Y)
    df['xbin'] = df['xbin'].apply(lambda x: x.mid)
    df['ybin'] = df['ybin'].apply(lambda x: x.mid)
    df = df.replace(0, np.NaN)

    df_temp = df.drop(['del_eta','del_phi'], axis=1)
    df_temp = df.groupby(['xbin', 'ybin']).agg('sum').reset_index()

    # df_temp['p_T'] = (df_temp['p_T']-df_temp['p_T'].min())/(df_temp['p_T'].max()-df_temp['p_T'].min())
    return df_temp
```

```python
def plot_img(x,y,z,name):

  z = z.values.reshape((len(x),len(y)))
  x,y  = np.array(x), np.array(y)
  fig, ax = plt.subplots()
  im = ax.imshow(z, extent=[x.min(), x.max(), y.min(), y.max()], 
                 origin='lower',cmap = 'jet', 
                 norm=matplotlib.colors.LogNorm())
  
  cbar = ax.figure.colorbar(im, ax=ax)
  cbar.set_label('sum value of p_T')
  
  plt.xlabel('Δη')
  plt.ylabel('ΔΦ')
  plt.title(name)
  plt.savefig(f'/Users/apple/Downloads/{name}.png')
  plt.show()
```
